```python
import pandas as pd
import numpy as np
```

```python
import liwc
import re
from collections import Counter
```

```python
# Carregar e formatar o dicionário 
def read_liwc_dictionary(filepath, encoding='ISO-8859-1'):
    with open(filepath, 'r', encoding=encoding) as file:
        lines = file.readlines()
    return lines


file_path = 'LIWC2007_Portugues_win.dic'
liwc_lines = read_liwc_dictionary(file_path)


temp_file_path = 'temp_liwc.dic'
with open(temp_file_path, 'w', encoding='utf-8') as temp_file:
    temp_file.writelines(liwc_lines)

parse, category_names = liwc.load_token_parser(temp_file_path)
```

```python
def tokenize(text):
    tokens = []
    
    for match in re.finditer(r'\w+', text, re.UNICODE):
        tokens.append(match.group(0))
    
    return tokens
```

```python
df = pd.read_csv('transcricoes_medium/m_transc_pe_arcanjo.csv', index_col=False)
```

```python
df = df.reindex(columns=df.columns.tolist() + category_names)
df[category_names] = 0
```

```python
for idx, row in df.iterrows():
    tokens = tokenize(row.Transcription)
    liwc_counts = Counter(
        category for token in tokens for category in parse(token)
    )

#     df.loc[idx, list(liwc_counts.keys())] = 1
    df.loc[idx, list(liwc_counts.keys())] = np.array(list(liwc_counts.values()))/row.preprocess_words
#     df.loc[idx, list(liwc_counts.keys())] = liwc_counts.values()
```

```python
# Significância 2 a 2 
from scipy.stats import chi2_contingency


def chi_square_test(df1, df2, padre1, padre2):
    results = []
    categories = df1['category'].tolist()
    
    for category in categories:
        if category in df2['category'].values:
            count1 = df1[df1['category'] == category]['count'].values[0]
            count2 = df2[df2['category'] == category]['count'].values[0]
            total1 = df1['count'].sum()
            total2 = df2['count'].sum()
            
            obs = np.array([
                [count1, count2],
                [total1 - count1, total2 - count2]
            ])
            
            chi2, p, dof, ex = chi2_contingency(obs, correction=False)
            
            if p <= 0.05:
                results.append({
                    'category': category,
                    'padre1': padre1,
                    'padre2': padre2,
                    'chi2': chi2,
                    'p_value': p,
                    'significance': 'yes'
                })
            else:
                results.append({
                    'category': category,
                    'padre1': padre1,
                    'padre2': padre2,
                    'chi2': chi2,
                    'p_value': p,
                    'significance': 'no'
                })
    
    return pd.DataFrame(results)

# Lista de arquivos CSV
files = [
    'results_liwc/result_m_transc_pe_arcanjo.csv',
    'results_liwc/results_m_transc_Bernardo_P_Küster.csv',
    'results_liwc/results_m_transc_Patrick_Fernandes.csv',
    'results_liwc/results_m_transc_pe.paulo.csv',
    'results_liwc/results_m_transc_Pe.Leonardo.csv',
    'results_liwc/results_transc_Leonardo_DallOsto.csv',
    'results_liwc/results_transc_Padre_Jose.csv'
]

# Carregar todos os CSVs em um dicionário de DataFrames
data = {file: pd.read_csv(file) for file in files}


result_folder = 'results_liwc/significance_results'
os.makedirs(result_folder, exist_ok=True)

# lista para armazenar todas as categorias significativas
significant_results = []

# Arcanjo contra os demais
arcanjo_df = data['results_liwc/result_m_transc_pe_arcanjo.csv']

for file, df in data.items():
    if file != 'results_liwc/result_m_transc_pe_arcanjo.csv':
        padre1 = 'result_m_transc_pe_arcanjo'
        padre2 = os.path.basename(file).replace('.csv', '')
        result_df = chi_square_test(arcanjo_df, df, padre1, padre2)
        result_file_path = os.path.join(result_folder, f'significance_arcanjo_vs_{os.path.basename(file)}')
        result_df.to_csv(result_file_path, index=False)
        significant_results.append(result_df)

# DallOsto contra os demais
dallosto_df = data['results_liwc/results_transc_Leonardo_DallOsto.csv']

for file, df in data.items():
    if file != 'results_liwc/results_transc_Leonardo_DallOsto.csv':
        padre1 = 'results_transc_Leonardo_DallOsto'
        padre2 = os.path.basename(file).replace('.csv', '')
        result_df = chi_square_test(dallosto_df, df, padre1, padre2)
        result_file_path = os.path.join(result_folder, f'significance_dallosto_vs_{os.path.basename(file)}')
        result_df.to_csv(result_file_path, index=False)
        significant_results.append(result_df)

# Comparação 2 a 2 para cada par de padres
for i, (file1, df1) in enumerate(data.items()):
    for j, (file2, df2) in enumerate(data.items()):
        if i < j:
            padre1 = os.path.basename(file1).replace('.csv', '')
            padre2 = os.path.basename(file2).replace('.csv', '')
            result_df = chi_square_test(df1, df2, padre1, padre2)
            result_file_path = os.path.join(result_folder, f'significance_{padre1}_vs_{padre2}.csv')
            result_df.to_csv(result_file_path, index=False)
            significant_results.append(result_df)

# Concatenar todos os resultados em um único DataFrame
final_df = pd.concat(significant_results, ignore_index=True)

# Filtrar apenas os resultados significativos
significant_df = final_df[final_df['significance'] == 'yes']

# Salvar os DataFrames finais em arquivos CSV
final_df.to_csv('results_liwc/significance_results/final_results.csv', index=False)
significant_df.to_csv('results_liwc/significance_results/significant_results_only.csv', index=False)

# Imprimir o DataFrame final
print("Teste de significância concluído e resultados salvos.")
print(final_df)
print("\nResultados significativos:")
print(significant_df)
```
